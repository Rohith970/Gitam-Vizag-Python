### Visualizing the data
- An interesting way to diplay DataFrame
- with matplotib

```python
import matplotlib.pyplot as plt
a1=[1,2,3,4]
a2=[5,6,7,8]
plt.figure(figsize=(5,5))
plt.plot(a1,a2,color='red')
plt.xlabel('Avg')
plt.ylabel('Sum')
plt.show()
```
